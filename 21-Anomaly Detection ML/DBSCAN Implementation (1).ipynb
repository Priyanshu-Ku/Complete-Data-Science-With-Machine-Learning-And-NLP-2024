## DBSCAN Clutering

```python
from sklearn.cluster import DBSCAN
## from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import make_circles ## New import for circular data
```

```python
X, y = make_circles(n_samples=750, factor=0.3, noise=0.1)  ## Changed to make_circles
```

```python
plt.scatter(X[:,0],X[:,1]) ## Plotting the circular data
plt.show()
```

```python
from sklearn.cluster import DBSCAN ## Importing DBSCAN clustering algorithm
```

```python
dbcan=DBSCAN(eps=0.1) ## Creating DBSCAN object with eps=0.1
```

```python
dbcan.fit_predict(X) ## Fitting the model and predicting the clusters
```

```python
dbcan.labels_ ## Getting the labels assigned by DBSCAN
```

```python
plt.scatter(X[:,0],X[:,1],c=dbcan.labels_) ## Plotting the clusters assigned by DBSCAN
plt.show()
```

```python
plt.scatter(X[:,0],X[:,1],c=y) ## Plotting the original labels
plt.show()
```
