## Local Outlier Factor (LOF) Anomaly Detection
The Local Outlier Factor (LOF) is an unsupervised anomaly detection algorithm that identifies outliers by measuring the local density deviation of a given data point with respect to its neighbors. It is particularly effective in detecting anomalies in datasets with varying densities.

```python
## Importing Libraries
from sklearn.datasets import make_circles ## to create a circular dataset
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
X,y=make_circles(n_samples=500,factor=0.5,noise=0.2,random_state=42) ## creating circular dataset
```

```python
## Visualizing the dataset
plt.scatter(X[:,0],X[:,1])  ## scatter plot 
plt.show()
```

```python
from sklearn.neighbors import LocalOutlierFactor ## importing LocalOutlierFactor Anomaly Detection model
```

```python
LOF = LocalOutlierFactor() ## creating object of LocalOutlierFactor Model
```

```python
LOF.fit_predict(X) ## fitting the model on data and predicting the outliers
```

```python
LOF.negative_outlier_factor_ ## getting the outlier factor values
```

```python
plt.scatter(X[:,0],X[:,1],c=LOF.negative_outlier_factor_) ## scatter plot with outlier factor values as color
plt.show()
```

```python
plt.scatter(X[:,0],X[:,1],c=y) ## scatter plot with original labels as color
plt.show()
```
