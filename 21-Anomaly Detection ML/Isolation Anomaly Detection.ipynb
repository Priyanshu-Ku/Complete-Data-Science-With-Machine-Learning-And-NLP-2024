## Isolation Forest Anomaly Detection

```python
import pandas as pd
df=pd.read_csv('healthcare.csv') ## Load the healthcare dataset
df.head()
```

```python
import matplotlib.pyplot as plt 
plt.scatter(df.iloc[:,0], df.iloc[:,1]) ## Scatter plot of the two features
plt.show()
```

```python
from sklearn.ensemble import IsolationForest ## Import Isolation Forest model
```

```python
clf = IsolationForest(contamination=0.2) ## Initialize the model
clf.fit(df) ## Fit the model to the data
predictions = clf.predict(df) ## Get predictions (-1 for anomalies, 1 for normal points)
```

```python
predictions
```

```python
import numpy as np
index = np.where(predictions < 0) ## Get indexes of anomalies
index
```

```python
x=df.values ## Convert dataframe to numpy array
```

```python
index = np.where(predictions < 0) 
plt.scatter(df.iloc[:,0], df.iloc[:,1]) ## Scatter plot of the two features
plt.scatter(x[index,0], x[index,1], edgecolors="r") ## Highlight anomalies
plt.show()
```
