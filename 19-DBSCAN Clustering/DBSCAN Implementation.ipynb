```python
## DBSCAN Clustering Example
from sklearn.cluster import DBSCAN ## Import DBSCAN from sklearn
from sklearn.datasets import make_moons ## Import make_moons to create a sample dataset that is not linearly separable 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
X,y=make_moons(n_samples=250,noise=0.05) ## Create dataset with 250 samples and some noise
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1]) ## Plot the dataset to visualize its structure, use c=y to color by true labels or get to see the output clusters
plt.show()
```

```python
##feature scaling(Standard Scaling)
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(X) ## Fitting and transforming the data
```

```python
X_scaled
```

```python
dbcan=DBSCAN(eps=0.3) ## Create DBSCAN object with eps=0.3
```

```python
dbcan.fit(X_scaled) ## Fit the DBSCAN model to the scaled data
```

```python
dbcan.labels_ ## Get the cluster labels assigned by DBSCAN
```

```python
plt.scatter(X[:,0],X[:,1],c=dbcan.labels_) ## Plot the original data colored by DBSCAN cluster labels
plt.show()
```

```python
plt.scatter(X[:,0],X[:,1],c=y) ## Plot the original data colored by true labels
plt.show()
```
