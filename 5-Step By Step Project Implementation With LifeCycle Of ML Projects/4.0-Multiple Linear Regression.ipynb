```python
## Lets prepare the dataframe 
dataset=pd.DataFrame(california.data,columns=california.feature_names)
dataset.head()
```

```python
dataset['Price']=california.target
```

```python
dataset.head()
```

```python
## Independent and Dependent features
X=dataset.iloc[:,:-1] #independent features
y=dataset.iloc[:,-1] #dependent features
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=10) 
## Splitting the data into training and testing
```

```python
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler ## Normalization
scaler=StandardScaler() ## Initialize the StandardScaler
```

```python
X_train=scaler.fit_transform(X_train) ## Fit and transform the X_train data
X_test=scaler.transform(X_test) ## Transform the X_test data
```

```python
X_train
```

```python
X_test
```

## Model Training

```python
from sklearn.linear_model import LinearRegression ## Import Linear Regression for Multiple Linear Regression
```

```python
regression=LinearRegression() ## Initialize the Linear Regression model
```

```python
regression
```

```python
regression.fit(X_train,y_train)
```

```python
## slope or coefficients
regression.coef_
```

```python
## intercepts
regression.intercept_
```

```python
## Prediction for the test data
y_pred=regression.predict(X_test)
```

```python
y_test
```

```python
y_pred
```

```python
## Performance Metrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
## R square and adjusted R square

from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
print(score)
```

```python
## Adjusted R square
#display adjusted R-squared
1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```

## Assumptions

```python
## 1st assumption the test data and predicted data should be linear
plt.scatter(y_test,y_pred)
plt.xlabel("Test Truth Data")
plt.ylabel("Test Predicted Data")
plt.show()
```

## Pickling 
Python pickle module is used for serialising and de-serialising a Python object structure. Any object in Python can be pickled so that it can be saved on disk. What pickle does is that it serialises the object first before writing it to file. Pickling is a way to convert a python object (list, dict, etc.) into a character stream. The idea is that this character stream contains all the information necessary to reconstruct the object in another python script.

```python
import pickle 
pickle.dump(regression,open('regressor.pkl','wb')) ## Save the model to disk
```

```python
model=pickle.load(open('regressor.pkl','rb')) ## Load the model from disk
```

```python
model.predict(X_test) ## Predict using the loaded model
```
